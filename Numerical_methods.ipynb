```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
# Moments of inertia around the principal axes (customize for your racquet)
I1 = 100  # Moment of inertia around stable axis (largest)
I2 = 40  # Moment of inertia around an unstable axis
I3 = 30  # Moment of inertia around 3rd axis

# Initial conditions
omega1 = 0.01  # Initial angular velocity around the stable axis
omega2 = 0.01  # Initial angular velocity around an unstable axis
omega3 = 1.0  # Initial angular velocity around 3rd axis

# Time settings
t0 = 0.0
t_end = 100.0
step_size = 0.1

# Lists to store time and angular velocity values
t_values = []
omega1_values = []
omega2_values = []
omega3_values = []

# Runge-Kutta 4th-order simulation
def calculate_alpha(omega1, omega2, omega3):
    alpha1 = ((I2 - I3) / I1) * omega2 * omega3
    alpha2 = ((I3 - I1) / I2) * omega1 * omega3
    alpha3 = ((I1 - I2) / I3) * omega1 * omega2
    return alpha1, alpha2, alpha3

def runge_kutta(omega1, omega2, omega3, step_size):
    k1_alpha1, k1_alpha2, k1_alpha3 = calculate_alpha(omega1, omega2, omega3)
    k2_alpha1, k2_alpha2, k2_alpha3 = calculate_alpha(omega1 + 0.5 * step_size * k1_alpha1,
                                                      omega2 + 0.5 * step_size * k1_alpha2,
                                                      omega3 + 0.5 * step_size * k1_alpha3)
    k3_alpha1, k3_alpha2, k3_alpha3 = calculate_alpha(omega1 + 0.5 * step_size * k2_alpha1,
                                                      omega2 + 0.5 * step_size * k2_alpha2,
                                                      omega3 + 0.5 * step_size * k2_alpha3)
    k4_alpha1, k4_alpha2, k4_alpha3 = calculate_alpha(omega1 + step_size * k3_alpha1,
                                                      omega2 + step_size * k3_alpha2,
                                                      omega3 + step_size * k3_alpha3)

    new_omega1 = omega1 + (step_size / 6) * (k1_alpha1 + 2 * k2_alpha1 + 2 * k3_alpha1 + k4_alpha1)
    new_omega2 = omega2 + (step_size / 6) * (k1_alpha2 + 2 * k2_alpha2 + 2 * k3_alpha2 + k4_alpha2)
    new_omega3 = omega3 + (step_size / 6) * (k1_alpha3 + 2 * k2_alpha3 + 2 * k3_alpha3 + k4_alpha3)

    return new_omega1, new_omega2, new_omega3

t = t0
while t < t_end:
    t_values.append(t)
    omega1_values.append(omega1)
    omega2_values.append(omega2)
    omega3_values.append(omega3)

    omega1, omega2, omega3 = runge_kutta(omega1, omega2, omega3, step_size)
    t += step_size

# Plot the results
plt.figure()
plt.plot(t_values, omega1_values, label='Angular Velocity 1')
plt.plot(t_values, omega2_values, label='Angular Velocity 2')
plt.plot(t_values, omega3_values, label='Angular Velocity 3')
plt.xlabel('Time')
plt.ylabel('Angular Velocity')
plt.title("Rotational Motion of a Tennis Racquet (Runge-Kutta 4)")
plt.legend()
plt.grid(True)
plt.show()
```
